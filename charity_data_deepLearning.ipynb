## Data Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("Resources/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME', which do not add any value in classification.

application_df1 = application_df.drop(["EIN", "NAME"], axis = 1)

```

```python
application_df1.head()
```

```python
# Determiing the number of unique values in each column.
application_df1.nunique()
```

```python
# looking at "APPLICATION_TYPE" value counts for binning purpose
app_types = application_df1["APPLICATION_TYPE"].value_counts()
app_types
```

```python
# Choose a cutoff value and create a list of application types to be replaced

application_types_to_replace = list(app_types[app_types<500].index)

# Replace above application types in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classif_count = application_df1["CLASSIFICATION"].value_counts()
classif_count
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classif_list = list(classif_count[classif_count>1])
classif_list
```

```python
#choosing cutoff value of 1000 for number of classification types and creating a list of classifications to be replaced
classifications_to_replace = list(classif_count[classif_count < 1000].index)


# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df1.head()
```

```python
#converting categorical data into numeric 
application_df2 = pd.get_dummies(application_df1)
application_df2
```

```python
#splitting the data into features and target arrays
y = application_df2["IS_SUCCESSFUL"]
X = application_df2.drop('IS_SUCCESSFUL', axis = 1)
X.shape
```

```python
#split the data into training and testing datasets with default 80/20 train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 40)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
#finding the number of features
print(X_train_scaled.shape)
print(X_train_scaled.shape[1])
```

```python
#defining the neural network model
no_of_features = X_train_scaled.shape[1]  #116 features
layer_1_nodes = 100
layer_2_nodes = 200
# layer_3_nodes = 50

activation_function = "relu"


nn = tf.keras.models.Sequential()
nn.add(tf.keras.layers.Dense(units=layer_1_nodes, activation = activation_function, input_dim=no_of_features))
nn.add(tf.keras.layers.Dense(units=layer_2_nodes, activation=activation_function))
# nn.add(tf.keras.layers.Dense(units=layer_3_nodes, activation=activation_function))

#adding output layer
nn.add(tf.keras.layers.Dense(units=1, activation = "sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=60)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
